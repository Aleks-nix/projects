### Замена названий столбцов и типов данных

```python
# Приведем названия столбцов к нижнему регистру
data_games.columns = data_games.columns.str.lower()
data_games
```

```python
# Выведем строки с незаполненным названием игры
data_games[data_games['name'].isnull() == True]
```

```python
# Удалим строки с незаполненным названием игры
data_games.dropna(subset = ['name'], inplace = True)
```

```python
# Выведем строки с пустым значением года релиза
data_games[data_games['year_of_release'].isnull() == True]
```

Видим что в колонке 'year_of_release' тип данных не числовой, это из-за 269 пустых значений, это менее 2-х процентов всех данных, кроме того заполнить год релиза в этих строках мы не сможем. Удалим эти строки.

```python
# Удаление строк с пропущенным годом релиза
data_games.dropna(subset = ['year_of_release',], inplace = True)
# Заменим вещественного типа данных на целочисленный
data_games['year_of_release'] = data_games['year_of_release'].astype('int')
data_games.info()
```

```python
# Найдем дубликаты по трем колонкам (название, платформа и год релиза)
data_games[data_games.duplicated(subset=['name', 'platform', 'year_of_release']) == True]
```

В данных есть один дубликат, удалим его

```python
data_games = data_games.drop_duplicates(subset=['name', 'platform', 'year_of_release']).reset_index(drop = True)
```

```python
data_games.info()
```

### Подготовка данных в колонках с оценкой критиков и пользователей и рейтингом

Пропуски в колонках 'critic_score' b 'user_score' заполним медианным значением по жанрам игр

```python
# Замена NaN в столбце 'critic_score' на ''
data_games['critic_score'] = data_games['critic_score'].fillna(value='')
# Замена NaN в столбце 'user_score' на ''
data_games['user_score'] = data_games['user_score'].fillna(value='')
```

В колонке с оценкой пользователей есть аббревиатура 'tbd', что расшифровывается как 'To Be Determined' и переводится "будет определено", но так как таблица не обновится и мы будем работать с этими значениями, то будем считать, что это пустые значения

```python
# Замена 'tbd' в столбце 'user_score' на ''
def del_tbd(row):
    if row == 'tbd':
        return ''
    else:
        return row
# Применение функции 'del_tbd' ко всем строкам
data_games['user_score'] = data_games['user_score'].apply(del_tbd)
```

```python
data_games.info()
```

```python
# data_games.info()
```

Мы заполнили пропуски в колонках с оценкой критиков и пользователей. В колонке с рейтингом от организации ESRB пропуски заполним значением 'undefined'

```python
# Замена NaN в столбце 'rating' на 'undefined'
data_games['rating'] = data_games['rating'].fillna(value='undefined')
```

```python
data_games.info()
```

### Игры разных платформ

Найдем суммарные продажи для каждой из платформ

```python
# Выведем таблицу суммарных продаж для каждой платформы в порядке убывания продаж
dg_platform_sales = data_games.pivot_table(index='platform',
                                           values='total_sales',
                                           aggfunc='sum').sort_values(by = 'total_sales', ascending = False)
```

```python
dg_platform_sales
```

Выберем платформы с наибольшими суммарными продажами, это первые 6 платформ, так как в 7-ой по счету платформе суммарные продажи упали вдвое по сравнению с предыдущей.

Построим распределение по годам для каждой из этих шести платформ.

```python
# Первые 6 платформ определим в список
columns = dg_platform_sales.index[0:6]
# Сгруппируем игры каждой платформы по годам, посчитаем количество игр и построим график распределения
for col in columns:
    data_games[data_games['platform'] == col].groupby('year_of_release')['platform'].count().plot(grid=True, 
                                                                                                  style='-o',
                                                                                                  label=col,
                                                                                                  figsize=(12, 5),
                                                                                                  legend=True)
    plt.xlabel('Год', size = 15)
    plt.ylabel('Количество', size = 15)
```

По графику видно, что время жизни платформы примерно 8-12 лет. Для прогнозирования данных на 2017 год и учитывая время жизни платформы, определим актуальный временной период, дальше будем рассматривать данные только с 2011 года!

```python
# Выберем актуальные данные
dg_actual = data_games.query('year_of_release >= 2011')
dg_actual
```

### Влияние отзывов на продажи игр на PS3

Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

```python
# Выведем данные по платформе PS3
dg_actual_critic_score = dg_actual.query('critic_score != ""')
dg_actual_critic_score['critic_score'] = dg_actual_critic_score['critic_score'].astype('float')
dg_actual_critic_score.query('platform == "PS3"')
```

```python
# Построим диаграмму рассеяния для продаж и отзывов критиков
print('Связь продаж и отзывов критиков')
dg_actual_critic_score.query('platform == "PS3"').plot(x='critic_score', y='total_sales', kind='scatter', alpha=0.3)
```

```python
# Найдем корреляцию между отзывами пользователей и продажами
dg_actual_user_score.query('platform == "PS3"')['total_sales'].corr(dg_actual_user_score.query('platform == "PS3"')['user_score'])
```

## Проверка гипотез
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.

Поясните:
- Как вы сформулировали нулевую и альтернативную гипотезы;
- Какой критерий применили для проверки гипотез и почему.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?

Формулировка гипотез:
- Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC равны;
- Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC не равны.

Критический уровень статистической значимости выберем равным 0.05, если результат сравнения окажется меньше него - отвергнем нулевую гипотезу.

```python
# Выделим значения пользовательского рейтинга для платформы Xbox One
dg_actual_ratio_xboxone = dg_actual_user_score.query('platform == "XOne"')['user_score']
dg_actual_ratio_xboxone
```

```python
# Выделим значения пользовательского рейтинга для платформы PC
dg_actual_ratio_pc = dg_actual_user_score.query('platform == "PC"')['user_score']
dg_actual_ratio_pc
```

```python
# Метод для проверки гипотезы для среднего двух выборок
alpha = 0.05
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(dg_actual_ratio_xboxone, dg_actual_ratio_pc)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Полученное значение p-value говорит о том, что хотя средний пользовательский рейтин для Xbox и отличается от среднего пользовательского рейтинга PC, с вероятностью в 62% такое или большее различие можно получить случайно. Это слишком большая вероятность, чтобы делать вывод о значимом различии между средними пользовательскими рейтингами.

### Средние пользовательские рейтинги жанров Action и Sports разные?

Формулировка гипотез:
- Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны;
- Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports не равны.

Критический уровень статистической значимости выберем равным 0.05, если результат сравнения окажется меньше него - отвергнем нулевую гипотезу.

```python
# Выделим значения пользовательского рейтинга для жанра Action 
dg_actual_genre_Action = dg_actual_user_score.query('genre == "Action"')['user_score']
dg_actual_genre_Action
```

```python
# Выделим значения пользовательского рейтинга для жанра Sports 
dg_actual_genre_Sports = dg_actual_user_score.query('genre == "Sports"')['user_score']
dg_actual_genre_Sports
```

```python
# Метод для проверки гипотезы для среднего двух выборок
alpha = 0.05
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(dg_actual_genre_Action, dg_actual_genre_Sports)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вероятность равенства средних пользовательских рейтингов жанров Action и Sports очень мала, а значит вероятность того, что они разные очень высока!
