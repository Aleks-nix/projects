## План проекта

Для построения модели, которая предскажет температуру стали, необходимо подготовить данные, на которых модель будет обучаться. Для этого выполним несколько задач:
- в данных об электродах для каждой партии найти суммарное время нагрева электродами, количество запусков нагрева электродами, суммарные активные и реактивные мощности, а так же можно посчитать полную мощность по формуле $$power=\sqrt{AP^2+RP^2},$$ где AP - 'active_power', а RP - 'reactive_power'
- для каждой партии найдем суммарную массу сыпучих и проволочных материалов и количество их подач
- из полученных данных сформировать таблицу признаков
- в качестве целевого признака использовать последнюю измеренную температу партии. Если такая температура отсутствует, то придется удалить данные всей партии. Первая температура не должна равняться последней. Первую температуру партии можно изпользовать в качестве одного из признаков, промежуточные измерения температуры нам не нужны
- проанализировать признаки на корреляционную зависимость
- обучить несколько моделей

## Подготовка данных

### Данные об электродах
Для каждой партии найдем суммарное время нагрева электродами, количество запусков нагрева электродами, суммарные активные и реактивные мощности, а так же можно посчитать полную мощность по формуле $$power=\sqrt{AP^2+RP^2},$$ где AP - 'active_power', а RP - 'reactive_power'

```python
data_arc.head()
```

```python
# Поменяем тип данных у дат
data_arc['start_heating'] = pd.to_datetime(data_arc['start_heating'])
data_arc['end_heating'] = pd.to_datetime(data_arc['end_heating'])
# Найдем время подогрева электродами в каждой сессии
data_arc['time_heating'] = data_arc['end_heating'] - data_arc['start_heating']
```

```python
data_arc.info()
```

```python
data_arc.head()
```

```python
# Для каждой партии найдем суммарные активные и реактивные мощности
pivot_arc = data_arc.pivot_table(index='key', values=['active_power', 'reactive_power'], aggfunc='sum')
```

```python
pivot_arc.head()
```

```python
# Для каждой партии найдем суммарное время нагрева
pivot_arc['time_heating'] = data_arc.groupby('key')['time_heating'].sum()
pivot_arc.head()
```

```python
# Переведем время нагрева в секунды
def seconds(row):
    return row.total_seconds()
pivot_arc['time_heating'] = pivot_arc['time_heating'].apply(seconds)
pivot_arc.head()
```

```python
# Найдем полную мощнось для каждой партии
pivot_arc['power'] = ((pivot_arc['active_power'])**2 + (pivot_arc['reactive_power'])**2)**0.5
```

```python
pivot_arc.head()
```

Эти данные дальше будем использовать в качестве признаков

### Данные о подаче материалов
для каждой партии найдем суммарную массу сыпучих и проволочных материалов и количество их подач

#### Сыпучие материалы

```python
data_bulk = round(data_bulk, 1)
data_bulk.head()
```

```python
# Напишем функции для нахождения суммарной массы материалов и подсчета подач материалов
def material_sum(row, columns):
    material_sum = 0
    for col in columns:
        material_sum += row[col]
    return material_sum

def material_count(row, columns):
    material_count = 0
    for col in columns:
        if row[col] != 0:
            material_count += 1
    return material_count
```

```python
bulk = ['Bulk 1', 'Bulk 2', 'Bulk 3', 'Bulk 4', 'Bulk 5', 'Bulk 6',
       'Bulk 7', 'Bulk 8', 'Bulk 9', 'Bulk 10', 'Bulk 11', 'Bulk 12',
       'Bulk 13', 'Bulk 14', 'Bulk 15']
```

```python
# Добавим в таблицу информацию о суммарной массе и количестве подач
data_bulk['bulk_sum'] = data_bulk.apply(material_sum, columns=bulk, axis=1)
data_bulk['bulk_count'] = data_bulk.apply(material_count, columns=bulk, axis=1)
```

```python
data_bulk.head()
```

```python
pivot_bulk = data_bulk[['key', 'bulk_sum', 'bulk_count']]
pivot_bulk.head()
```

#### Проволосные материалы

```python
data_wire = round(data_wire, 1)
data_wire.head()
```

```python
wire = ['Wire 1', 'Wire 2', 'Wire 3', 'Wire 4', 'Wire 5', 'Wire 6',
       'Wire 7', 'Wire 8', 'Wire 9']
```

```python
# Добавим в таблицу информацию о суммарной массе и количестве подач
data_wire['wire_sum'] = data_wire.apply(material_sum, columns=wire, axis=1)
data_wire['wire_count'] = data_wire.apply(material_count, columns=wire, axis=1)
```

```python
data_wire.head()
```

```python
pivot_wire = data_wire[['key', 'wire_sum', 'wire_count']]
pivot_wire.head()
```

```python
pivot_material = pivot_bulk.merge(pivot_wire, on='key')
pivot_material.head()
```

### Результаты измерения температур

```python
data_temp
```

```python
# Заменим все пропущенные значения температуры нулями
# В дальнейшем удалим значения всех партий, где температура = 0
data_temp = data_temp.fillna(value=0)
```

```python
# Для каждой партии возьмем первое время измерения температуры и температуру
pivot_first_temp = data_temp.pivot_table(index='key',values=['measurement_time', 'temp'],aggfunc='first')
pivot_first_temp
```

```python
# Для каждой партии возьмем последнее время измерения температуры и температуру
pivot_last_temp = data_temp.pivot_table(index='key',values=['measurement_time', 'temp'],aggfunc='last')
pivot_last_temp
```

```python
# Найдем все партии, где значение температуры было пропущено
pivot_last_temp.query('temp == 0')
```

```python
# Удалим этим партии
pivot_last_temp = pivot_last_temp.query('temp != 0')
pivot_last_temp
```

```python
# Объеденим данные по совпадающим значениям партий
pivot_temp = pivot_first_temp.merge(pivot_last_temp, on='key')
pivot_temp.columns = ['measurement_first_time','first_temp', 'measurement_last_time', 'last_temp']
pivot_temp
```

```python
# Создадим функцию, которая присваивает каждой партии 0, если последняя температура выше первой
# и присваивает 1 в остальных случаях
def error(row):
    if row['last_temp'] > row['first_temp']:
        return 0
    else:
        return 1
```

```python
# Применим функцию к каждой строке таблицы
pivot_temp['error'] = pivot_temp.apply(error, axis=1)
```

```python
# Выведем все партии, где error = 1
pivot_temp.query('error == 1')
```

Слишком много партий, где последняя температура меньше либо равна первой, поэтому будем считать, что это допустимо и оставим этим данные.

## Построение моделей

```python
# Выделим переменные признаки и признак, который нужно предсказать для каждой таблицы
features = pivot_data.drop('last_temp', axis=1)
target = pivot_data['last_temp']
# Отделим валидационную выборку (25%) от первоначальных данных
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=11111)
```

### Линейная регрессия

```python
model_lin_reg = LinearRegression()
result_lin_reg = cross_val_score(model_lin_reg, features_train, target_train, verbose=5,
                                 scoring='neg_mean_absolute_error', cv=5).mean()
print('MAE модели линейной регрессии:', abs(result_lin_reg))
```

```python
# Проверим модель линейной регрессии на тестовой выборке
model_lin_reg = LinearRegression()
model_lin_reg.fit(features_train, target_train)
predictions_test = model_lin_reg.predict(features_test)
MAE = mean_absolute_error(target_test, predictions_test)
print('MAE модели линейной регрессии на тестовой выборке:', MAE)
```

### LightGBM

```python
model_lgbm = LGBMRegressor(random_state=12345)
params_lgbm = {'n_estimators' : [100, 300, 500, 1000],
               'max_depth' : range(2, 7),
               'num_leaves': [20, 30, 40],
               'learning_rate': [0.03, 0.1, 0.5]}

search_lgbm = GridSearchCV(estimator=model_lgbm, param_grid=params_lgbm, scoring='neg_mean_absolute_error',
                           verbose=0, cv=5)
result_lgbm = search_lgbm.fit(features_train , target_train)
```

```python
print('MAE модели LightGBM:', abs(result_lgbm.best_score_))
print('Лучшие параметры LightGBM:', result_lgbm.best_params_)
```

```python
# Проверим модель LightGBM на тестовой выборке
model = result_lgbm.best_estimator_
model.fit(features_train, target_train)
predictions_test = model.predict(features_test)
MAE = mean_absolute_error(predictions_test, target_test)
print('MAE модели LightGBM на тестовой выборке:', MAE)
```

### XGBoost

```python
model_xgb = XGBRegressor(random_state=12345)
params_xgb = {'eta' : [0.01,0.1, 0.3, 0.5],
               'max_depth' : range(2, 10)}

search_xgb = GridSearchCV(estimator=model_xgb, param_grid=params_xgb, scoring='neg_mean_absolute_error',
                           verbose=0, cv=5)
result_xgb = search_xgb.fit(features_train , target_train)
```

```python
print('MAE модели XGBoost:', abs(result_xgb.best_score_))
print('Лучшие параметры XGBoost:', result_xgb.best_params_)
```
