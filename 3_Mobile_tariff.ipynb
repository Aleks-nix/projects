### Заполнение нулевых значений в звонках и интернет-сессиях

Вспомним, что компания всегда округляет вверх значения минут и мегабайтов, поэтому все нулевые значения в звонках и интернет трафике заменим на 1, а все ненулевые значения округлим в большую сторону.

```python
# Создадим функцию, заменяющую 0 на 1 в строках
def replace_0_1(row,):
    if row == 0:
        return 1
    else:
        return row
```

```python
data_calls['duration'] = data_calls['duration'].apply(replace_0_1)
# Вновь посчитаем количество строк с нулевой продолжительностью звонков
len(data_calls.query('duration == 0'))
```

```python
data_internet['mb_used'] = data_internet['mb_used'].apply(replace_0_1)
# Вновь посчитаем количество строк, в которых mb_used = 0
len(data_internet.query('mb_used == 0'))
```

### Округление в большую сторону значений в звонках и интернет-сессиях

```python
# Создадим фунцию, округляющую числа в большую сторону
def rounding(row):
    return math.ceil(row)
```

```python
# Применим фунцию округления в данных со звонками и вывод первых пяти строк
data_calls['duration'] = data_calls['duration'].apply(rounding)
data_calls.head()
```

### Подсчет сделанных звонков и израсходованных минут

```python
# Создание колонки с месяцем в таблице со звонками пользователей
data_calls['month'] = pd.to_datetime(data_calls['call_date'], format='%Y-%m-%d').dt.month
```

```python
# Создание сводной таблицы с количеством звонков по месяцам для каждого пользователя
data_calls_pivot = data_calls.pivot_table(index=['user_id', 'month'], values='duration', aggfunc=['count', 'sum'])
# Переименуем название стобцов
data_calls_pivot.columns = ['count_calls', 'sum_minutes']
# Вывод пяти первых строк сводной таблицы
data_calls_pivot.head()
```

### Подсчет отправленных сообщений

```python
# Создание колонки с месяцем в таблице с отправленными сообщениями
data_messages['month'] = pd.to_datetime(data_messages['message_date'], format='%Y-%m-%d').dt.month
```

```python
# Создание сводной таблицы с количеством отправленных сообщений по месяцам для каждого пользователя
data_messages_pivot = data_messages.pivot_table(index=['user_id', 'month'], values='id', aggfunc='count')
# Переименуем название стобцов
data_messages_pivot.columns = ['count_messages']
# Вывод пяти первых строк сводной таблицы
data_messages_pivot.head()
```

### Подсчет объема израсходованного интернет-трафика

```python
# Создание колонки с месяцем в таблице с объемом израсходованного интернет-трафика
data_internet['month'] = pd.to_datetime(data_internet['session_date'], format='%Y-%m-%d').dt.month
```

```python
# Создание сводной таблицы с объемом израсходованного интернет-трафика по месяцам для каждого пользователя
data_internet_pivot = data_internet.pivot_table(index=['user_id', 'month'], values='mb_used', aggfunc='sum')
# Переименуем название стобцов
data_internet_pivot.columns = ['sum_mb']
# Вывод пяти первых строк сводной таблицы
data_internet_pivot.head()
```

### Подсчет помесячной выручки с каждого пользователя
Вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану.

Для начала объединим все данные.

```python
# Объединим сводные таблицы с информацией о звонках, сообщениях и интернет-сессиях в таблицу data_all
data_all = data_calls_pivot.join(data_messages_pivot, on=['user_id', 'month'], how= 'outer')
data_all = data_all.join(data_internet_pivot, on=['user_id', 'month'], how= 'outer')
```

```python
# Сброс индексов
data_all = data_all.reset_index()
data_all
```

```python
# Выделяем колонки с пользователем и тарифом в отдельную таблицу
data_users_tariff = data_users.loc[:, ['user_id', 'tariff']]
data_users_tariff.head()
```

```python
# Создание функции, возвращающей название тарифа каждому пользователю
def tariff(row):
    if row in list(data_users_tariff.query('tariff == "ultra"')['user_id']):
        return 'ultra'
    if row in list(data_users_tariff.query('tariff == "smart"')['user_id']):
        return 'smart'
    else:
        return ''
```

```python
# Применение функции tariff к каждой строке таблицы data_all
data_all['tariff'] = data_all['user_id'].apply(tariff)
```

```python
# Заменим Nan на нули
data_all = data_all.fillna(value=0)
data_all
```

```python
# Созданем функции, которая в каждой строке вычитает бесплатный лимит из суммарного количества
# звонков, сообщений и интернет-трафика. И если эти значения больше нуля, то остаток умножают на значения
# из тарифного плана, к результату прибавляется абонентская плата соответственного тарифа.
def monthly_proceeds(row):
    if row['tariff'] == 'smart':
        difference_calls = row['sum_minutes'] - 500
        difference_messages = row['count_messages'] - 50
        difference_mb = row['sum_mb'] - 15360
        if difference_calls <= 0 and difference_messages <= 0 and difference_mb <= 0:
            decision = 550
            return decision
        if difference_calls > 0 and difference_messages <= 0 and difference_mb <= 0:
            decision = 550 + difference_calls*3
            return decision
        if difference_calls > 0 and difference_messages > 0 and difference_mb <= 0:
            decision = 550 + difference_calls*3 + difference_messages*3
            return decision
        if difference_calls > 0 and difference_messages > 0 and difference_mb > 0:
            decision = 550 + difference_calls*3 + difference_messages*3 + math.ceil(difference_mb/1024)*200
            return decision
        if difference_calls > 0 and difference_messages <= 0 and difference_mb > 0:
            decision = 550 + difference_calls*3 + math.ceil(difference_mb/1024)*200
            return decision
        if difference_calls <= 0 and difference_messages > 0 and difference_mb > 0:
            decision = 550 + difference_messages*3 + math.ceil(difference_mb/1024)*200
            return decision
        if difference_calls <= 0 and difference_messages <= 0 and difference_mb > 0:
            decision = 550 + math.ceil(difference_mb/1024)*200
            return decision
        if difference_calls <= 0 and difference_messages > 0 and difference_mb <= 0:
            decision = 550 + difference_messages*3
            return decision
        else:
            return 'error_smart'
    
    if row['tariff'] == 'ultra':
        difference_calls = row['sum_minutes'] - 3000
        difference_messages = row['count_messages'] - 100
        difference_mb = row['sum_mb'] - 30720
        if difference_calls <= 0 and difference_messages <= 0 and difference_mb <= 0:
            decision = 1950
            return decision
        if difference_calls > 0 and difference_messages <= 0 and difference_mb <= 0:
            decision = 1950 + difference_calls*1
            return decision
        if difference_calls > 0 and difference_messages > 0 and difference_mb <= 0:
            decision = 1950 + difference_calls*1 + difference_messages*1
            return decision
        if difference_calls > 0 and difference_messages > 0 and difference_mb > 0:
            decision = 1950 + difference_calls*1 + difference_messages*1 + math.ceil(difference_mb/1024)*150
            return decision
        if difference_calls > 0 and difference_messages <= 0 and difference_mb > 0:
            decision = 1950 + difference_calls*1 + math.ceil(difference_mb/1024)*150
            return decision
        if difference_calls <= 0 and difference_messages > 0 and difference_mb > 0:
            decision = 1950 + difference_messages*1 + math.ceil(difference_mb/1024)*150
            return decision
        if difference_calls <= 0 and difference_messages <= 0 and difference_mb > 0:
            decision = 1950 + math.ceil(difference_mb/1024)*150
            return decision
        if difference_calls <= 0 and difference_messages > 0 and difference_mb <= 0:
            decision = 1950 + difference_messages*1
            return decision
        else:
            return 'error_ultra'
```

```python
# Применим фунцию monthly_proceeds к каждой строке data_all
data_all['monthly_proceeds'] = data_all.apply(monthly_proceeds, axis=1)
```

## Проверка гипотез
- средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается

Формулировка гипотез:
- Нулевая гипотеза: средняя выручка пользователей «Ультра» равна средней выручке пользователей «Смарт»;
- Альтернативная гипотеза: средняя выручка пользователей «Ультра» не равна средней выручке пользователей «Смарт».

```python
# Выведем таблицу с помесячной оплатой за услуги мобильного оператора каждого пользователя
data_all.head()
```

```python
# Сформируем таблицу средней выручки каждого пользователя тарифа "Ультра"
data_all_proceeds_ultra = data_all.query('tariff == "ultra"').pivot_table(index='user_id',
                                                values='monthly_proceeds', aggfunc='mean')
data_all_proceeds_ultra
```

```python
# Сформируем таблицу средней выручки каждого пользователя тарифа "Смарт"
data_all_proceeds_smart = data_all.query('tariff == "smart"').pivot_table(index='user_id',
                                                values='monthly_proceeds', aggfunc='mean')
data_all_proceeds_smart
```

```python
# Метод для проверки гипотезы для среднего двух выборок
alpha = 0.05
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(data_all_proceeds_ultra, data_all_proceeds_smart)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Видим, что очень малая вероятность того, что средние выручки пользователей тарифа "Смарт" и "Ультра" равны, следовательно мы отвергаем нулевую гипотезу.

### Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Формулировка гипотез:
- Нулевая гипотеза: средняя выручка пользователей из Москвы равна выручке пользователей из других регионов;
- Альтернативная гипотеза: средняя выручка пользователей из Москвы не равна выручке пользователей из других регионов.

```python
# Выделяем колонки с пользователем и городом в отдельную таблицу
data_users_city = data_users.loc[:, ['user_id', 'city']]
data_users_city.head()
```

```python
# Создание функции, возвращающей название города каждому пользователю
def tariff(row):
    if row in list(data_users_city.query('city == "Москва"')['user_id']):
        return 'Москва'
    else:
        return 'Другой'
```

```python
# Применение функции tariff к каждой строке таблицы data_all
data_all['city'] = data_all['user_id'].apply(tariff)
```

```python
data_all
```

```python
# Сформируем таблицу средней выручки каждого пользователя из Москвы
data_all_proceeds_city_moscow = data_all.query('city == "Москва"').pivot_table(index='user_id',
                                                values='monthly_proceeds', aggfunc='mean')
data_all_proceeds_city_moscow
```

```python
# Сформируем таблицу средней выручки каждого пользователя из других регионов
data_all_proceeds_city_other = data_all.query('city == "Другой"').pivot_table(index='user_id',
                                                values='monthly_proceeds', aggfunc='mean')
data_all_proceeds_city_other
```

```python
# Метод для проверки гипотезы для среднего двух выборок
alpha = 0.05
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(data_all_proceeds_city_moscow, data_all_proceeds_city_other, equal_var = False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Полученное значение p-value говорит о том, что хотя средняя выручка пользователей и Москвы отличается от средней выручки пользователей из других регионов, с вероятностью в 35% такое или большее различие можно получить случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними выручками.
