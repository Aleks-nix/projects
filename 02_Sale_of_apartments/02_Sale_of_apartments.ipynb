## Изучение и предобработка данных

```python
import pandas as pd

# Код для считывания данных
data_sf = pd.read_csv('sale_flat.csv', sep='\t')
data_sf.head()
```

```python
data_sf.info()
```

Пропуски присутствуют в столбцах:
- ceiling_height (высота потолков)
- floors_total (всего этажей в доме)
- living_area (жилая площадь)
- is_apartment (апартаменты)
- kitchen_area (площадь кухни)
- balcony (число балконов)
- locality_name (название населённого пункта)
- airports_nearest (расстояние до ближайшего аэропорта)
- cityCenters_nearest (расстояние до центра города)
- parks_around3000 (число парков в радиусе 3 км)
- parks_nearest (расстояние до ближайшего парка)
- ponds_around3000 (число водоёмов в радиусе 3 км)
- ponds_nearest (расстояние до ближайшего водоёма)
- days_exposition (сколько дней было размещено объявление (от публикации до снятия))

Высота потолков, количество этажей в доме, жилая площадь, площадь кухни, название населенного пункта, расстояние до аэропорта. Пропуски в этих колонках не сможем заполнить, поэтому они остаются неизменными.

Пропуски в столбцах с апартемантами и балконом заполним отсутствием статуса апартаментов (False) и отсутствием балкона (0).

Число парков и водоемов частитчно можно заполнить, анализируя расстояние до ближайшего парка или водоема, если там есть информация (если ближайшая точка находится на расстоянии более 3000 м, то ставим 0)
Расстояние до ближайшего парка или водоема заполнить не сможем, оставим без изменений.

Скорее всего количество дней размещения объявления заполняется автоматически, если имеется пропуск, то объявление по каким-либо причинам еще размещено. Создадим новую колонку с информацией о размещении объявления на сайте (булев тип).

### Пропуски в столбцах с апартаментами и балконами

```python
# Поиск уникальных значений в столбце с апартаментами
data_sf['is_apartment'].unique()
```

```python
# Поиск уникальных значений в столбце с балконами
data_sf['balcony'].unique()
```

```python
# Замена NaN в апартаментах на False
data_sf['is_apartment'] = data_sf['is_apartment'].fillna(value=False)
data_sf['is_apartment'].unique()
```

```python
# Замена NaN в балконах на 0
data_sf['balcony'] = data_sf['balcony'].fillna(value=0)
data_sf['balcony'].unique()
```

### Пропуски в столбцах с парками

```python
# Поиск строк, где количество парков не 0 или не NaN.
data_sf.query('parks_around3000 != 0 and parks_around3000 != "NaN"')
```

```python
# Поиск строк где расстояние до ближайшего парка < 3000 м
data_sf.query('parks_nearest <= 3000')
```

```python
data_sf.query('parks_nearest > 3000')
```

В строках где есть наличие парков в радиусе 3000 м также присутствует информация о растоянии до ближайшего парка. Кроме этих строк есть еще 4 с информацией о ближайшем парке, но расстояние больше 3000, и в этих строках столбцы с количеством парков заполнены нулями. Следовательно значения 'parks_around3000' во всех строках с пропуском в колонке 'parks_nearest' можно заполнить нулями.

```python
# Заполняем все пропуски нулями
data_sf['parks_around3000'] = data_sf['parks_around3000'].fillna(value=0)
```

### Пропуски в столбцах с водоемами

```python
# Поиск строк, где количество водоемов не 0 или не NaN.
data_sf.query('ponds_around3000 != 0 and ponds_around3000 != "nan"')
```

```python
# Поиск строк где расстояние до ближайшего водоема < 3000 м
data_sf.query('ponds_nearest <= 3000')
```

Расстояние до ближайшего водоема указываектся в тех строках, где есть информация о водоемах в радиусе 3000 м. Следовательно можно заполнить столбец 'ponds_around3000' во всех остальных строках нулями.

```python
# Заполняем все пропуски нулями
data_sf['ponds_around3000'] = data_sf['ponds_around3000'].fillna(value=0)
data_sf.info()
```

### Пропуски сроках публикации объявлений

```python
# Заменим в столбце 'days_exposition' пропуски на '' и присвоим значения в новый столбец 'days_exposition_new'
data_sf['days_exposition_new'] = data_sf['days_exposition'].fillna(value='')
# Создание функции присваивания булевых значений
def timing_announcement(row):
    if row == '':
        return False
    else:
        return True
# Создание нового столбца, где True - есть сроки размещения объявления, False - нет сроков
data_sf['timing_announcement'] = data_sf['days_exposition_new'].apply(timing_announcement)
data_sf.head()
```

### Изменение типов данных

В столбцах с количеством этажей в доме, балконов, парков, водоемов и дней размещения обявлений можно заменить тип данных на целочисленный. Однако в столбцах с количеством этажей в доме и дней размещения объявлений присутствуют пропуски, поэтому мы можем заменить тип данных только в оставшихся колонках.

```python
columns_int = ['balcony', 'parks_around3000', 'ponds_around3000']
data_sf[columns_int] = data_sf[columns_int].astype('int')
```

```python
data_sf.info()
```

## Дополнение таблицы расчетными значениями 

Посчитайте и добавьте в таблицу:
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# Подсчет цены за квадратный метр
data_sf['price_square_meter'] = data_sf['last_price']/data_sf['total_area']
```

```python
# Создание колонки с днем недели публикации
data_sf['weekday'] = pd.to_datetime(data_sf['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S').dt.weekday
# Создание колонки с месяцем публикации
data_sf['month'] = pd.to_datetime(data_sf['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S').dt.month
# Создание колонки с годом публикации
data_sf['year'] = pd.to_datetime(data_sf['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S').dt.year
```

```python
# Создадим функцию, определяющую один из трех вариантов этажа: первый, последний, другой
def floor_definition(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
# Применение функции floor_definition ко всем строкам
data_sf['floor_definition'] = data_sf.apply(floor_definition, axis=1)
```

```python
# Создание колонки с отношением жилой площади к общей
data_sf['ratio_area_lt'] = data_sf['living_area'] /data_sf['total_area'] 
# Создание колонки с отношением площади кухни к общей площади
data_sf['ratio_area_kt'] = data_sf['kitchen_area'] /data_sf['total_area'] 
```

```python
# Вывод первых пяти строк получившейся таблицы
data_sf.head()
```

### Редкие и выбивающиеся значения
Отфильтруем данные, оставим объявления удовлетворяющие следующим условиям:
- 15 < площадь квартиры < 200 квадратных метров;
- 400 т.р. < цена на момент снятия объявления < 30 млн. р.;
- количество комнат < 8;
- 1.5 м < высота потолков < 5 м;
- время продажи < 1000 дней.

Отфильтрованной таблице присвоим новое название data_sf_good

```python
data_sf_good = data_sf.query('15 < total_area < 200 and 400 < last_price < 3e+07 and rooms < 8 and (1.5 < ceiling_height < 5 or ceiling_height == "nan") and (days_exposition < 1000 or days_exposition == "nan")')
```

```python
data_sf_good.shape
```

```python
# Построим гистограммы для выбранных параметров
columns = ['total_area', 'last_price', 'rooms', 'ceiling_height', 'days_exposition']
for col in columns:
    data_sf_good.hist(col, bins=50)
```

### Влияние некоторых факторов на стоимость квартиры

Определим как следующие параметры влияют на стоимость квартиры:
- площдь;
- количество комнат;
- удаленность от центра города;
- этаж квартиры (первый,последний, другой);
- дата размещения объявления (день недели, месяц, год).

```python
# Посмотрим как зависит смтоимость квартиры от площади, колличества комнат и расстояния до центра
columns = ['total_area', 'rooms', 'cityCenters_nearest']
for col in columns:
    print('Коэффициент корреляции между last_price и', col, '=', data_sf_good[col].corr(data_sf_good['last_price']))
    data_sf_good.plot(x=col, y='last_price', kind='scatter', figsize=(6, 4), alpha=0.1)
```

Видим, что цена квартиры зависит от количества комнат, однако от площади зависит сильнее. А так же видим, чем дальше от центра находится квартира тем меньше стоимость.

```python
# Посмотрим как зависит смтоимость квартиры от этажа, на котором она находится
pivot_table_floor = data_sf_good.pivot_table(index='floor_definition',
                                             values='last_price', aggfunc=['count', 'mean', 'median'])
pivot_table_floor.columns = ['count', 'mean', 'median']
pivot_table_floor
```

Квартиры с первых этажей самые дешевые, больше всего ценятся квартиры между первым и последним этажом.

```python
# Посмотрим как зависит смтоимость квартиры от даты размещения объявления
columns = ['weekday', 'month', 'year']
for col in columns:
    print('Коэффициент корреляции между last_price и', col, '=',
          data_sf_good[col].corr(data_sf_good['last_price']))
    data_sf_good.plot(x=col, y='last_price', kind='scatter', figsize=(6, 4), alpha=0.1)
```

```python
# Построим распределение количества продаваемых квартир от года
data_sf_good.pivot_table(index='year', values='last_price', aggfunc='count').plot(grid=True, figsize=(12, 5))
```

Дата размещения объявления никак не влияет на цену. Однако видно, что в 2017 и 2018 году количество продаваемых квартир больше чем в 2015, 2016 и 2019 года
