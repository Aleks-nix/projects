### Заполнение пропусков в колонке с трудовым стажем

Для заполнения пропусков в колонке 'days_employed' сгруппируем данные по возрасту клиента. Определим следующие группы:
- меньше 30 лет
- 30 - 40 лет
- 40 - 50 лет
- 50 - 65 лет
- больше 65 лет
Далее найдем средний стаж для каждой группы и заменим пропуски в соответсвии с возрастом клиента.

```python
# Создание функции распределения клиентов по возрастным группам
def group_age(age):
    if 18 <= age <= 30:
        return 'меньше 30'
    if 30 < age <= 40:
        return '30 - 40 лет'
    if 40 < age <= 50:
        return '40 - 50 лет'
    if 50 < age <= 65:
        return '50 - 65 лет'
    if 65 < age <= 100:
        return 'больше 65 лет'
    else: 
        return 'некорректно'
```

```python
# Применение функции group_age(age) на всех строках, кроме строк с пропусками в столце 'days_employed'
data_bc['group_age'] = data_bc['dob_years'].apply(group_age)
data_bc.head(15)
```

```python
# Удалим данные 10-ти клиентов
data_bc = data_bc[(data_bc['days_employed'] != '') |
                  ((data_bc['days_employed'] == '') & (data_bc['group_age'] != 'некорректно'))]
data_bc.info()
```

```python
# Найдем средний трудовой стаж в каждой группе
mean_before30 = data_bc[(data_bc['group_age'] == 'меньше 30') &
                        (data_bc['days_employed'] != '')]['days_employed'].mean()
mean_30_40 = data_bc[(data_bc['group_age'] == '30 - 40 лет') &
                        (data_bc['days_employed'] != '')]['days_employed'].mean()
mean_40_50 = data_bc[(data_bc['group_age'] == '40 - 50 лет') &
                        (data_bc['days_employed'] != '')]['days_employed'].mean()
mean_50_65 = data_bc[(data_bc['group_age'] == '50 - 65 лет') &
                        (data_bc['days_employed'] != '')]['days_employed'].mean()
mean_after65 = data_bc[(data_bc['group_age'] == 'больше 65 лет') &
                        (data_bc['days_employed'] != '')]['days_employed'].mean()

# Функция, заменяющая пропуски в 'days_employed' на среднее значение трудового стажа в зависимости
# от возрастной группы клиента
def mean_days_employed(row):
    if row['days_employed'] == '':
        if row['group_age'] == 'меньше 30':
            return mean_before30
        if row['group_age'] == '30 - 40 лет':
            return mean_30_40
        if row['group_age'] == '40 - 50 лет':
            return mean_40_50
        if row['group_age'] == '50 - 65 лет':
            return mean_50_65
        if row['group_age'] == 'больше 65 лет':
            return mean_after65
    else:
        return row['days_employed']
```

```python
# Применение функции mean_days_employed(row) ко всем строкам data_bc
data_bc['days_employed'] = data_bc.apply(mean_days_employed, axis=1)
```

### Заполнение пропусков в колонке с доходом

Для заполнения пропусков в колонке 'total_income' сгруппируем данные по типу занятости и найдем медианное значение в каждой группе.

```python
# Выделим все типы занятости
data_bc['income_type'].unique()
```

```python
# Поиск медианного значения дохода при каждом типе занятости
median_employee = data_bc[(data_bc['income_type'] == 'сотрудник') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_retiree = data_bc[(data_bc['income_type'] == 'пенсионер') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_companion = data_bc[(data_bc['income_type'] == 'компаньон') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_civil_servant = data_bc[(data_bc['income_type'] == 'госслужащий') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_unemployed = data_bc[(data_bc['income_type'] == 'безработный') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_businessman = data_bc[(data_bc['income_type'] == 'предприниматель') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_student = data_bc[(data_bc['income_type'] == 'студент') &
                          (data_bc['total_income'] != '')]['total_income'].median()
median_decree = data_bc[(data_bc['income_type'] == 'в декрете') &
                          (data_bc['total_income'] != '')]['total_income'].median()

# Функция, заменяющая пропуски в 'total_income' на медианное значение дохода в зависимости типа занятости клиента
def median_total_income(row):
    if row['total_income'] == '':
        if row['income_type'] == 'сотрудник':
            return median_employee
        if row['income_type'] == 'пенсионер':
            return median_retiree
        if row['income_type'] == 'компаньон':
            return median_companion
        if row['income_type'] == 'госслужащий':
            return median_civil_servant
        if row['income_type'] == 'безработный':
            return median_unemployed
        if row['income_type'] == 'предприниматель':
            return median_businessman
        if row['income_type'] == 'студент':
            return median_student
        if row['income_type'] == 'в декрете':
            return median_decree
    else:
        return row['total_income']
```

```python
# Применение функции median_total_income(row) ко всем строкам data_bc
data_bc['total_income'] = data_bc.apply(median_total_income, axis=1)
```

### Зависимость между наличием детей и возвратом кредита в срок

```python
# Разобьем клиентов на группы в зависимости от количесвтва детей и сосчитаем клиентов в каждой группе
data_bc.groupby('children')['children'].count()
```

Напишем функцию, которая будет создавать сводную таблицу, ее аргументом будет столбец, по значениям которого происходит группировка. Она будет показывать количество клиентов, которые имели ('1') или не имели ('0') задолженность перед банком, а так же подсчитывать отношение ('ratio') клиентов ('1'/'0'). После этого  полученные данные сортируются по убыванию 'ratio'.

```python
# Функция, создающая сводную таблицу
def pivon_table(column):
    # Создание сводной таблицы
    data_bc_pivot = data_bc.pivot_table(index=[column], columns='debt',
                                        values='days_employed', aggfunc='count')
    # Замена пропусков на '0'
    data_bc_pivot= data_bc_pivot.fillna(value='0')
    # Переводим в числовой тип
    data_bc_pivot = data_bc_pivot.astype('int')
    # Находим отношение
    data_bc_pivot['ratio'] = data_bc_pivot[1]/data_bc_pivot[0]
    # Сортируем данные по убыванию 'ratio'
    return data_bc_pivot.sort_values(by = 'ratio', ascending = False)
```

```python
pivon_table('children')
```

Вывод: Скорее всего клиенты, которые указали 20 детей, ошиблись и данные на их основе не стоит рассматривать. Клиентов с 5 детьми мало, поэтому этими данными тоже можно пренебречь. Отсюда видно, что самый низний процент должников среди клиентов без детей и клиентов с тремя детьми.

### Зависимость между уровнем дохода и возвратом кредита в срок

```python
# Найдем минимальный и максимальный доход
total_income_min = data_bc['total_income'].min()
total_income_max = data_bc['total_income'].max()
print('Минимальный доход клиентов =', total_income_min)
print('Максимальный доход клиентов =', total_income_max)
```

Разброс дохода огромный. Разобьем доходы клиентов на следующие группы:
- 20т - 40т
- 40т - 70т
- 70т - 100т
- 100т - 150т
- 150т - 300т
- 300т - 500т
- 500т - 1млн
- свыше 1млн

```python
# Функция, определеющая доход клиента к одной из предложенных групп
def group_total_income(income):
    if income <= 40000:
        return '20т - 40т'
    if 40000 < income <= 70000:
        return '40т - 70т'
    if 70000 < income <= 100000:
        return '70т - 100т'
    if 100000 < income <= 150000:
        return '100т - 150т'
    if 150000 < income <= 300000:
        return '150т - 300т'
    if 300000 < income <= 500000:
        return '300т - 500т'
    if 500000 < income <= 1000000:
        return '500т - 1млн'
    if 1000000 < income:
        return 'свыше 1млн'
    else: 
        return 'ошибка'
```

```python
# Применяем функцию к каждой строке таблицы
data_bc['group_total_income'] = data_bc['total_income'].apply(group_total_income)
```

```python
# Подсчет клиентов в каждой группе
data_bc.groupby('group_total_income')['group_total_income'].count()
```

```python
pivon_table('group_total_income')
```

Сюда же отнеслись и клиенты с доходом в 40-70 тысяч, вероятно это ответственные люди, которые берут кредиты исходя из своего дохода и справляются с выплатами.

### Зависимость между целями кредита и его возвратом в срок

```python
# Разбиение по группам и подсчет клиентов
data_bc.groupby('purpose')['purpose'].count()
```

Различных целей много, но можно выделить 4 основные группы:
- покупка автомобиля
- получение образования
- операции с недвижимостью
- свадьба

```python
# Функция, определеющая цель получения кредита к одной из 4-х групп
def group_purpose(lemmas):
    if 'автомобиль' in lemmas:
        return 'автомобиль'
    if 'образование' in lemmas:
        return 'образование'
    if ('недвижимость' in lemmas) or  ('жилье' in lemmas):
        return 'недвижимость'
    if 'свадьба' in lemmas:
        return 'свадьба'
    else:
        return 'ошибка'
```

```python
# Применяем функцию к каждой строке таблицы
data_bc['group_purpose'] = data_bc['lemmas'].apply(group_purpose)
```
